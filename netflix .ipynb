<h1 style = "color : #e50914"> Data Pre-Processing</h1>


```python
#dropping the coloumn of description 
#dropping a coloumn of show id 
#changing the data type of date added and release year 
```

```python
netflix.drop('description' , axis = 1 , inplace = True)
```

```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'])
```

```python
netflix['year_added'] = netflix['date_added'].dt.year
```

```python
netflix.drop('date_added' , axis = 1 , inplace = True)
```

```python
netflix.drop('cast' , axis = 1 , inplace = True)
```

```python
netflix[netflix['director'].isnull()]
```

```python
netflix.loc[netflix['director'].isnull(),'director'] = 'Unknown'
```

```python
netflix['director'].isnull().sum()
```

```python
netflix.loc[netflix['country'].isnull(),'country'] = 'Unknown'
```

```python
netflix['country'].isnull().sum()
```

```python
netflix
```

```python
netflix['director'].nunique()
```

```python
netflix['first_country'] = netflix['country'].apply(lambda x: x.split(",")[0])
```

## Movies and TV shows added over time 

```python
dt = netflix.groupby('year_added')['show_id'].count().sort_values( ascending = False)
```

```python
sns.lineplot(dt, color='#b20710',linewidth=3)
plt.xlabel('Year')
plt.ylabel('Number of titles')
plt.title('Netflix evolution')
plt.grid()
plt.show()
```
